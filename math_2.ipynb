## Объединение множеств

При использовании на 3 множествах получаем новый объект, содержащий элементы 3 х множеств (без повторов). Операция объединения выполним двумя способам: с помощью символа | или метода union()

```python
set_1 = {1, 2, 3}
set_2 = {2, 3, 4, 5}
set_3 = {6, 7, 8}
U = set_1 | set_2 | set_3
print(U)
# используя метод union
U = set_1.union(set_2, set_3)
print(U)
```

## Разность. Симметричная разность

В функцию diff() передаем  в качестве аргументов 3 исходных множества и дополнительный аргумент функции symmetric, которая будет возвращать либо симметричную разность, либо просто разность (если дополнительный аргумент функции symmetric имеет значение False)

Разность двух множеств показывает уникальные элементы первого, которых нет во втором.
Симметричная разность выводит уникальные компоненты обоих контейнеров, исключая общие для них. Решение можно провести как с помощью операторов, так и методов.

Вариант 1. При помощи операторов

```python
def diff(set_1, set_2, set_3, symmetric=True):
    if symmetric:
        return set_1 ^ set_2 ^ set_3
    return set_1 - set_2 - set_3
```

```python
set_1 = {3, 4, 5, 6, 20}
set_2 = {4, 6, 7, 8, 9}
set_3 = {5, 3, 8, 1}

print(diff(set_1, set_2, set_3))
print(diff(set_1, set_2, set_3, symmetric=False))
```

Вариант 2. При помощи методов

```python

def diff(set_1, set_2, set_3, symmetric=True):
    if symmetric:
        return set_1.symmetric_difference(set_2).symmetric_difference(set_3)
    return set_1.difference(set_2, set_3)

set_1 = {3, 4, 5, 6, 20}
set_2 = {4, 6, 7, 8, 9}
set_3 = {5, 3, 8, 1}

print(diff(set_1, set_2, set_3))
print(diff(set_1, set_2, set_3, symmetric=False))
```
